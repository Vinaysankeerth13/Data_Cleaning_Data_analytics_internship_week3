```python
data.columns = data.columns.str.lower().str.replace(' ','_')
```

# Age

```python
avg_age = data.age.mean()
```

```python
data['age'] = data['age'].fillna(avg_age)
data['age'] = data['age'].round(decimals=1)
data['age']
```

```python
data['age'] = data['age'].astype(int)
data['age']
```

```python
data['age'].isna().sum()
```

# Salary

```python
data['salary'] = data['salary'].apply(lambda x:x.replace('$',''))
```

```python
data['salary']
```

```python
data['salary'] = data['salary'].apply(lambda x:x.replace('k',''))
```

```python
data['salary']
```

# Rating

```python
rating_avg = data['rating'].mean()
data['rating'] = data['rating'].fillna(rating_avg)
data['rating']
```

```python
data.info()
```

#Location

```python
data.head()
```

```python
data['location'] = data['location'].apply(lambda x:x.replace('India,In','India'))
data['location'] = data['location'].apply(lambda x:x.replace('New York,Ny','New York'))
data['location'] = data['location'].apply(lambda x:x.replace('Australia Aus','Australia'))
data['location']
```

#Established

```python
data['established']
```

```python
data['established'] = data['established'].apply(lambda x: x if x > 0 else pd.NA)
data['established']
```

#Easy_Apply

```python

data['easy_apply']
```

```python
data['easy_apply'] = data['easy_apply'].apply(lambda x:x.replace('-1','FALSE'))
data['easy_apply'] = data['easy_apply'].map({'TRUE': True, 'FALSE': False})
data['easy_apply']
```
